```python
# Separate target before encoding if you prefer 
# But I will encode only feature categorical cols (exclude target if it's object)
cat_feature_cols = [c for c in cat_cols if c != target]
print("Categorical feature columns to encode:", cat_feature_cols)
```

```python
# One-hot encode categorical features (drop_first=True to avoid dummy trap)
df_encoded = pd.get_dummies(df, columns=cat_feature_cols, drop_first=True)
```

```python
# Map target to numeric if needed
if df_encoded[target].dtype == object:
    df_encoded[target] = df_encoded[target].map({'yes': 1, 'no': 0})
    if df_encoded[target].isnull().any():
        print("Warning: target mapping produced NaNs; unique values:", df[target].unique())
```

```python
#  Define X and y
X = df_encoded.drop(columns=[target])
y = df_encoded[target].astype(int)

print("Encoded dataframe shape:", df_encoded.shape)
print("Features shape:", X.shape)
print("Target distribution:\n", y.value_counts())
```
